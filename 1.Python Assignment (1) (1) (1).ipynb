<h3> Q1: Given two matrices please print the product of those two matrices </h3>
<pre>

Ex 1: A   = [[1 3 4]
             [2 5 7]
             [5 9 6]]
      B   = [[1 0 0]
             [0 1 0]
             [0 0 1]]
      A*B = [[1 3 4]
             [2 5 7]
             [5 9 6]]

     
Ex 2: A   = [[1 2]
             [3 4]]
      B   = [[1 2 3 4 5]
             [5 6 7 8 9]]
      A*B = [[11 14 17 20 23]
             [23 30 36 42 51]]
             
Ex 3: A   = [[1 2]
             [3 4]]
      B   = [[1 4]
             [5 6]
             [7 8]
             [9 6]]
      A*B =Not possible
</pre>

```python
# Function for the matrix multiplication
def mat_mul(A, B):
    row_A = len(A)
    columns_A = len(A[0])
    rows_B = len(B)
    columns_B = len(B[0])

    # To multiply  m×n matrix by an n×p matrix, the ns must be the same,
    #  the result is an m×p matrix.
    if columns_A != rows_B:
        print(
            "ERROR!.....Matrix multiplication of two arguments not possible as number of columns in first Matrix is NOT equal to the number of rows in second Matrix")
        return


    # result matrix will have dimensions of rows_A x columns_B
    # And fill the matrix with zeros
    result = [[0 for i in range(columns_B)] for j in range(row_A)]

    # Actual Logic for multiplication
    for i in range(row_A):
        for j in range(columns_B):
            for k in range(columns_A):
                # k-th column of A should be the k-th row of B
                result[i][j] += A[i][k] * B[k][j]
    return result
    #Matrix given are
A = [[1, 3, 4],
     [2, 5, 7],
     [5, 9, 6]]

B = [[1, 0, 0],
     [0, 1, 0],
     [0, 0, 1]]

A1 = [[1, 2],
      [3, 4]]

B1 = [[1, 2, 3, 4, 5],
      [5, 6, 7, 8, 9]]
A2 = [[1, 3, 4], [5, 9, 6]]

B2 = [[1, 0, 0], [0, 0, 1]]
#Display the multiplication results
print(mat_mul(A, B))
print(mat_mul(A1, B1))
print(mat_mul(A2, B2))
```

<h3> Q3: Replace the digits in the string with #</h3>

consider a string that will have digits in that, we need to remove all the not digits and replace the digits with #
<pre>
Ex 1: A = 234                Output: ###
Ex 2: A = a2b3c4             Output: ###
Ex 3: A = abc                Output:   (empty string)
Ex 5: A = #2a$#b%c%561#      Output: ####
</pre>

```python
import re
# Replace and remove all non digit  values
def replace_remove_non_digits(str):
    #Values for the replacement
    r = [
        ('\D', '',),
        ('\d', '#'),
    ]

    for o, n in r:
        str = re.sub(o, n, str)

    return str

# cases
s1 = "a2b3c4"
s2 = "234"
s3 = 'abc'
s4 = '#2a$#b%c%561#'
#print the output
print(replace_remove_non_digits(s1))
print(replace_remove_non_digits(s2))
print(replace_remove_non_digits(s3))
print(replace_remove_non_digits(s4))
```

<h3> Q5: Find the closest points</h3>

consider you have given n data points in the form of list of tuples like S=[(x1,y1),(x2,y2),(x3,y3),(x4,y4),(x5,y5),..,(xn,yn)] and a point P=(p,q) <br> your task is to find 5 closest points(based on cosine distance) in S from P
<br>cosine distance between two points (x,y) and (p,q) is defind as $cos^{-1}(\frac{(x\cdot p+y\cdot q)}{\sqrt(x^2+y^2)\cdot\sqrt(p^2+q^2)})$
<pre>
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1)(6,0),(1,-1)]
P= (3,-4)
<img src='https://i.imgur.com/vIFPOcG.jpg', width=300>
Output:
(6,-7)
(1,-1)
(6,0)
(-5,-8)
(-1,-1)
</pre>

```python


import math
# here S is list of tuples and P is a tuple of len=2
cos_list = []
S = [(1,2), (3,4), (-1,1), (6,-7), (0, 6), (-5,-8), (-1,-1), (6,0), (1,-1)]
P = (3,-4)

# To find out the inverse of cosine in Python we use math.acos() function of Python Standard math Library. 
# The inverse of cosine is also called arc cosine.
for a, b in S:
  #logic for finding closet point
    nm = a * P[0] + b * P[1]
    dm = math.sqrt(a ** 2 + b ** 2) * math.sqrt(P[0] ** 2 + P[1] **2)       
    cos_list.append(math.acos(nm / dm))
#printing the outputs
X = cos_list
Y = [S for S in sorted(zip(S,X), key = lambda i: i[1])]
k = Y[:5] 
for i, j in k:
    print(i)
```

<h3> Q6: Find Which line separates oranges and apples</h3>
consider you have given two set of data points in the form of list of tuples like 
<pre>
Red =[(R11,R12),(R21,R22),(R31,R32),(R41,R42),(R51,R52),..,(Rn1,Rn2)]

<pre>

Red= [(1,1),(2,1),(4,2),(2,4), (-1,4)]
Blue= [(-2,-1),(-1,-2),(-3,-2),(-3,-1),(1,-3)]
Lines=["1x+1y+0","1x-1y+0","1x+0y-3","0x+1y-0.5"]
<img src='https://i.imgur.com/DoQf7mE.jpg' width=400>
Output:

NO

</pre>

```python
import math

def one(red, blue, line):
    sign_red = -1

    if eval(line.replace('x', '*%s' % red[0][0]).replace('y', '*%s' % red[0][1])) > 0:
        sign_red = 1
    #the  set of two traversal for the blue-points tuples
    for red_pt in red:
        if sign_red == 1 and eval(
                line.replace('x', '*%s' % red_pt[0]).replace('y', '*%s' % red_pt[1])) < 0:
            return 'NO'
            #compute blue point tuples
        if sign_red == -1 and eval(
                line.replace('x', '*%s' % red_pt[0]).replace('y', '*%s' % red_pt[1])) > 0:
            return 'NO'

    # Now the same set of two traversal for the blue-points tuples
    sign_blue = -1 * sign_red

    for blue_pt in blue:
        if sign_blue == 1 and eval(
                line.replace('x', '*%s' % blue_pt[0]).replace('y', '*%s' % blue_pt[1])) < 0:    
                #compute blue point tuples
            return 'NO'

        if sign_blue == -1 and eval(
                line.replace('x', '*%s' % blue_pt[0]).replace('y', '*%s' % blue_pt[1])) > 0:
            return 'NO'

    return 'YES'


Red = [(1, 1), (2, 1), (4, 2), (2, 4), (-1, 4)]
Blue = [(-2, -1), (-1, -2), (-3, -2), (-3, -1), (1, -3)]

Lines = ["1x+1y+0", "1x-1y+0", "1x+0y-3", "0x+1y-0.5"]
#printing the outputs
for i in Lines:
    yes_or_no = one(Red, Blue, i)
    print(yes_or_no)
```

<h3> Q7: Filling the missing values in the specified formate</h3>
You will be given a string with digits and '\_'(missing value) symbols you have to replace the '\_' symbols as explained 
<pre>
Ex 1: _, _, _, 24 ==> 24/4, 24/4, 24/4, 24/4 i.e we. have distributed the 24 equally to all 4 places <br>
Ex 2: 40, _, _, _, 60 ==> (60+40)/5,(60+40)/5,(60+40)/5,(60+40)/5,(60+40)/5 ==> 20, 20, 20, 20, 20 i.e. the sum of (60+40) is distributed qually to all 5 places<br>
Ex 3: 80, _, _, _, _  ==> 80/5,80/5,80/5,80/5,80/5 ==> 16, 16, 16, 16, 16 i.e. the 80 is distributed qually to all 5 missing values that are right to it<br>
Ex 4: _, _, 30, _, _, _, 50, _, _  
==> we will fill the missing values from left to right 
    a. first we will distribute the 30 to left two missing values (10, 10, 10, _, _, _, 50, _, _)
    b. now distribute the sum (10+50) missing values in between (10, 10, 12, 12, 12, 12, 12, _, _) 
    c. now we will distribute 12 to right side missing values (10, 10, 12, 12, 12, 12, 4, 4, 4)
</pre>
for a given string with comma seprate values, which will have both missing values numbers like ex: "_, _, x, _, _, _"
you need fill the missing values

Q: your program reads a string like ex: "_, _, x, _, _, _" and returns the filled sequence

Ex: 
<pre>
Input1: "_,_,_,24"
Output1: 6,6,6,6

Input2: "40,_,_,_,60"
Output2: 20,20,20,20,20

Input3: "80,_,_,_,_"
Output3: 16,16,16,16,16

Input4: "_,_,30,_,_,_,50,_,_"
Output4: 10,10,12,12,12,12,4,4,4
</pre>



```python
def curve_smoothing(string):
   
    list = []
#split string , is applied
    SplitStr = string.split(',')
    
    for i in range(len(SplitStr)):
        if SplitStr[i] != '_':
           list.append(i)
            
    # add the length of the original list to list
    list.append(len(SplitStr) - 1)
    
    # Loop over the list to modify the original list by filling up empty slots
    
    s = 0
    for j in list:
        # Whenever there is a missing value, add the proceeding value and the immediately next available value
        # and divide it by the number of missing values.
        # Calculate the cumulative_sum of available non-empty cells.
        value = int(SplitStr[j]) if SplitStr[j] != '_' else 0
        value += int(SplitStr[s]) if SplitStr[s] != '_' and s != j else 0

        # Divide the cumulative_sum of prev and next value by the number of missing values
        replace = value // (j - s + 1)
        SplitStr = [replace if s <= x <= j else SplitStr[x] for x in range(len(SplitStr))]
        s = j
    return SplitStr     #list of values

S =  "_,_,30,_,_,_,50,_,_"
smoothed_values= curve_smoothing(S)
print(smoothed_values)
```

Your task is to find the value of $f(Y,Y_{score}) = -1*\frac{1}{n}\Sigma_{for each Y,Y_{score} pair}(Ylog10(Y_{score})+(1-Y)log10(1-Y_{score}))$ here n is the number of rows in the matrix
<pre>
Ex:
[[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
output:
0.4243099
</pre>
$\frac{-1}{8}\cdot((1\cdot log_{10}(0.4)+0\cdot log_{10}(0.6))+(0\cdot log_{10}(0.5)+1\cdot log_{10}(0.5)) + ... + (1\cdot log_{10}(0.8)+0\cdot log_{10}(0.2)) )$

```python
from math import log
#imort log from math liabrary
def log_loss(A):
    crossentropy = 0   #entropy initialize to zero
    for row in A:
        #compute cross entropy
          crossentropy += (row[0] * log(row[1], 10) + ((1 - row[0]) * log(1 - row[1], 10)))
#compute log loss
    log_loss = -1 *  crossentropy / len(A)
    return log_loss

#print the output
A = [[1, 0.4], [0, 0.5], [0, 0.9], [0, 0.3], [0, 0.6], [1, 0.1], [1, 0.9], [1, 0.8]]
print(log_loss(A))
```
